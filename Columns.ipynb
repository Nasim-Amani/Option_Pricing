```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "alldata_fi.csv"
df = pd.read_csv(file_path)

# Convert Date column to datetime and sort
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by=['Date'])

# Select relevant columns
columns_needed = ['Date', 'option_name', 'Closing_Price_(Option)', 'Exercise_Price',
                  'Spot_Price_of_Asset', 'riskfreerate', 'time_to_expiration']
df = df[columns_needed]

# Set up plot style
sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))

# Plot each numerical feature over time
numeric_columns = ['Closing_Price_(Option)', 'Exercise_Price',
                   'Spot_Price_of_Asset', 'riskfreerate', 'time_to_expiration']

for i, col in enumerate(numeric_columns, 1):
    plt.subplot(len(numeric_columns), 1, i)
    plt.plot(df['Date'], df[col], label=col)
    plt.xlabel('Date')
    plt.ylabel(col)
    plt.title(f'{col} over Time')
    plt.legend()

plt.tight_layout()
plt.show()
```
